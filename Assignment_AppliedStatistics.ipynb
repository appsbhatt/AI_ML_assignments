```python
# do the charges of the people who smoke differ significantly from people who dont
mtpl.bar(insur['smoker'], insur['charges']) 
mtpl.xlabel("Smoker") 
mtpl.ylabel("Charges")

# Yes, it does differ as the charges of the non smokers is less than that of the smokers.
```

```python
# does bmi of males differ significantly from that of females
mtpl.bar(insur['sex'], insur['bmi']) 
mtpl.xlabel("Sex") 
mtpl.ylabel("BMI")

# Not much difference but in males the BMI tends to exceed  45 in males where  in female it doesnt
```

```python
#is the distribution of bmi across women with no children, one child and two children, the same? 

# 1.making a new dataframe where it is ensured that only values of females are there.
newdataframe = insur[(insur['sex']=='female')&(insur['children']<=3)]

# 2.printing the new dataframe
newdataframe.head()
```

```python
# 3.plotting the bmi vs no. children plot
sns.catplot(x="children", y="bmi", jitter=False, data=newdataframe)

# According to the plot which we have made we can see it is very evident that the distribution in all the cases is almost same.
```

```python
# is the proportion of smokers significantly different in different genders?

# changing the categorical value(non numeric) for sex into numeric value. (1 for male and 2 for female)
insur.sex[insur.sex=='male']=1
insur.sex[insur.sex=='female']=2
insur.head() 

```

```python
# visualising the data according to the values
mtpl.bar(insur['smoker'], insur['sex']) 
mtpl.xlabel("Smoker") 
mtpl.ylabel("sex")

# It is the same proportion in both the genders.
```
