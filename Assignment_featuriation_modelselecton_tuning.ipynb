```python
for i in concrete_df_zscore.columns:
    q1, q2, q3 = concrete_df_zscore[i].quantile([0.25,0.5,0.75])
    IQR = q3 - q1
    a = concrete_df_zscore[i] > q3 + 1.5*IQR
    b = concrete_df_zscore[i] < q1 - 1.5*IQR
    concrete_df_zscore[i] = np.where(a | b, q2, concrete_df_zscore[i])  
    ##The outliers are relatively treated  with interquartile range
```

```python
#the score is really not good for kmeans clustering so we will drop this

#Now we will perform PCA to check whether that can improve performance or not

from sklearn.decomposition import PCA


cov_matrix = np.cov(clus2_z, rowvar=False)

np.linalg.eig(cov_matrix)

eigenvalues, eigenvectors, = np.linalg.eig(cov_matrix)

eig_pairs = [(eigenvalues[index], eigenvectors[:,index]) for index in range(len(eigenvalues))]

# Sort the (eigenvalue, eigenvector) pairs from lowest to highest with respect to eigenvalue
eig_pairs.sort()
eig_pairs.reverse()    # reverses the sorted pairs from increasing value of eigenvalue to lowest

# Extract the descending ordered eigenvalues and eigenvectors
eigvalues_sort = [eig_pairs[index][0] for index in range(len(eigenvalues))]
eigvectors_sort = [eig_pairs[index][1] for index in range(len(eigenvalues))]
```

```python
tot = sum(eigenvalues)

var_explained = [(i / tot) for i in sorted(eigenvalues, reverse=True)]  
cum_var_exp = np.cumsum(var_explained)  


cum_var_exp
```

```python
plt.bar(range(0, 9), var_explained, alpha=0.5, align='center', label='individual explained variance')
plt.step(range(0,9),cum_var_exp, where= 'mid', label='cumulative explained variance')
plt.ylabel('Explained variance ratio')
plt.xlabel('Principal components')
plt.legend(loc = 'best')
plt.show()
```
